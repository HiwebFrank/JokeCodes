## 酒量 - 跑量 简单关系模型

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DrinkQty = np.arange(0, 10, 0.1)
RunningDistance = -DrinkQty + 10
```

```python
plt.xlabel("Drink Qty (Liang)")     # 酒量 单位：两
plt.ylabel("Running Distance (KM)")  # 跑量 单位：公里
plt.title("DrinkQty-RunningDistance Relation Model")

plt.plot(DrinkQty, RunningDistance)
```
